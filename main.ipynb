```python
from classes.food_dataframe import FoodDataFrame
from classes.food_charts import FoodCharts


if __name__ == "__main__":
    while True:
        num_barcodes = input("How many products would you like to compare?")
        if not num_barcodes.isdigit():
            print("\nInvalid number. Please, enter a number.")
        else:
            num_barcodes = int(num_barcodes)
            break 

    food_dataframe: FoodDataFrame = FoodDataFrame(num_barcodes)
    food_charts: FoodCharts = FoodCharts(food_dataframe.create_dataframe())
    
    food_dataframe.display_table()
    food_charts.plot_data()
```
